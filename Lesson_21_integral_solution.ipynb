## Метод прямоугольников (численное интегрирование)

```python
from math import *
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 5 + 3/x + np.sin(x)

def F(x):
    return 5*x + 3*np.log(x) - np.cos(x)

a = 2.
b = 10.
N = 8
h = (b - a) / N

# интеграл методом прямоугольников
s = 0
for i in range(N):
    s += f(a + i*h)
s = s * h
print('Интеграл от', a, 'до', b, '=', s)
print('Ошибка метода', abs(F(b) - F(a) - s))
print("Относительная ошибка метода:", abs(F(b) - F(a) - s) / s)

# рисуем график данной функции:
diapasone1 = np.arange(a, b+h, h)
diapasone2 = np.arange(a, b+h/100, h/100)
plt.plot(diapasone1, f(diapasone1), 'ro')
plt.plot(diapasone2, f(diapasone2))
plt.show()
 
```

## Метод трапеций(численное интегрирование)

```python
from math import *
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 5 + 3/x + np.sin(x)

def F(x):
    return 5*x + 3*np.log(x) - np.cos(x)

a = 2.
b = 10.
N = 8
h = (b - a) / N

# интеграл методом трапеций
s = (f(b) - f(a)) * h / 2
for i in range(N):
    s += f(a + i*h)
s = s * h
print('Интеграл от', a, 'до', b, '=', s)
print('Ошибка метода', abs(F(b) - F(a) - s))
print("Относительная ошибка метода:", abs(F(b) - F(a) - s) / s)

# рисуем график данной функции:
diapasone1 = np.arange(a, b+h, h)
diapasone2 = np.arange(a, b+h/100, h/100)
plt.plot(diapasone1, f(diapasone1), 'ro')
plt.plot(diapasone2, f(diapasone2))
plt.show()
```

## Метод Симпсона

```python
from math import *
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 5 + 3/x + np.sin(x)

def F(x):
    return 5*x + 3*np.log(x) - np.cos(x)

a = 2.
b = 10.
N = 100
h = (b - a) / N

# интеграл методом Симпсона
s = f(a) + f(b)
for i in range(1, N // 2):
    s += 2 * f(a + 2*i*h)
for i in range(0, N // 2):
    s += 4 * f(a + (2*i + 1)*h)
s = s * h / 3
print('Интеграл от', a, 'до', b, '=', s)
print('Ошибка метода', abs(F(b) - F(a) - s))

# рисуем график данной функции:
diapasone1 = np.arange(a, b+h, h)
diapasone2 = np.arange(a, b+h/100, h/100)
plt.plot(diapasone1, f(diapasone1), 'ro')
plt.plot(diapasone2, f(diapasone2))
plt.show()
```
